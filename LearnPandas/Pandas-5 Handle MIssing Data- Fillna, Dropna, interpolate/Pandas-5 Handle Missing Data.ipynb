```python
import pandas as pd
df = pd.read_csv("weather_data.csv", parse_dates=['day'])
# type(df.day[0])
df.set_index('day', inplace=True)
df
```

# Fill Na 
using missing valu handeling

```python
new_df = df.fillna(0)
new_df
```

## Specific column values using fillna


```python
new_df = df.fillna({
    'temperature' : 0.0,
    'windspeed' : 0,
    'event' : 'NoEvent'
})
new_df
```

## fill na using forward values

```python
new_df = df.fillna(method='ffill')  #also using backword -> bfill
new_df
```

## Interpolate

```python
df
```

```python
new_df = df.interpolate()
new_df
```

## interpolate with some parrameters

```python
new_df = df.interpolate(method = "time")
new_df
```

## Drop Na  Values rows


```python
df
```

```python
 # if the full cell values are Na then remove
new_df = df.dropna(how='all')
new_df
```

```python
# drop cell based on some conditions
new_df = df.dropna(thresh=2)
new_df
```

## incerting the missing dates

```python
dt = pd.date_range("01-01-2017", "01-11-2017")
idx = pd.DatetimeIndex(dt)
df.reindex(idx)
```

```python
# after adding all the dates fill the Na values using interpolating
new_df = df.interpolate(method = "time")
new_df
```
